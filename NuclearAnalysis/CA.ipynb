<h6 style="font-size: 40px; color: #1E54C8;">
    What is Nuclear Energy?</h6>

## Short Brief

![Nuclear Image](nuclearimg.jpg)


###  Nuclear energy is a form of energy released from the nucleus, the core of atoms, made up of protons and neutrons. This source of energy can be produced in two ways: fission when nuclei of atoms split into several parts or fusion when nuclei fuse together (Colando, 2022)

### Uranium ore is obtained through mining the earth, it goes through the milling process to extract the uranium from the stone and ore, this process usually happens on site where the mining happens where it gets milled into a uranium oxide concnetrate called yellowcake due to its yellow appearance. Enrichment process happens where it seperates gaseous uranium hexafluoride into two streams, one being enriched to the required level known as low-enriched uranium (LEU) and the other where it gets depleted (tails). More processes of enrichment is followed upon such as gas centrifuge enrichment and laser seperation.

The atom splitting is typically initiated by bombarding a heavy atom, such as uranium-235 or plutonium-239, with a neutron. When the nucleus of the atom absorbs the neutron, it becomes unstable and splits into two smaller nuclei, along with a few free neutrons and a large amount of energy in the form of heat. ## Imports

```python
import numpy as np
import pandas as pd #to show the dataset
import matplotlib.pyplot as plt
```

### This first example will analyse which year and month produced the most net energy 

### Dataset read

```python
sample_data = pd.read_csv('nuclear_energy_view.csv')  # linking the dataset using pandas
sample_data  # showcasing it
```

```python
print(sample_data.columns) #showcasing the coulmns of the dataset
```

### Grouping the year and month together 

```python
yearly_generation = sample_data.groupby("Year")["Nuclear Electricity Net Generation"].sum()
monthly_generation = sample_data.groupby("Month")["Nuclear Electricity Net Generation"].sum()
```

### Finding the highest net generation from each year and month 


```python
peak_year = yearly_generation.idxmax(), yearly_generation.max()
peak_month = monthly_generation.idxmax(), monthly_generation.max() 

print("Peak Month:", peak_month[0])
print("Maximum Generation in Peak Month across all years:", peak_month[1])

print("Peak Year:", peak_year[0])
print("Maximum Generation in Peak Year:", peak_year[1])
```

### As we can see from the data the most production for nuclear energy happens in Janurary new years resolution you can say, before it dwindles down 

#### Now lets see which Janurary year produced the most energy.

```python
# Filtering data for January
january_data = sample_data[sample_data["Month"] == "January"]

# Group by Year and calculate the total generation for each January
january_generation = january_data.groupby("Year")["Nuclear Electricity Net Generation"].sum()

# Find the year with the maximum generation in January
peak_january_year = january_generation.idxmax()
peak_january_generation = january_generation.max()

# Printing the results
print("Year with Highest January Generation:", peak_january_year)
print("Maximum Generation in January:", peak_january_generation)
```

### Overall monthly and yearly production

```python
# Sort yearly and monthly generation
yearly_generation_sort = yearly_generation.sort_values(ascending=False)
monthly_generation_sort = monthly_generation.sort_values(ascending=False)

print("Yearly Generation:\n", yearly_generation_sort)
print("Monthly Generation:\n", monthly_generation_sort)
```

### Visual Representation 

```python
import matplotlib.pyplot as plt

# Bar graph yearly generation
plt.figure(figsize=(12, 6))
yearly_generation.plot(kind='bar', color='skyblue')
plt.title('Yearly Nuclear Electricity Net Generation')
plt.xlabel('Year')
plt.ylabel('Net Generation (GWh)')
plt.show()

# Bar graph monthly generation
plt.figure(figsize=(10, 6))
monthly_generation.plot(kind='bar', color='green')
plt.title('Average Monthly Nuclear Electricity Net Generation')
plt.xlabel('Month')
plt.ylabel('Net Generation (GWh)')
plt.show()
```

## What can we tell from the data and graphs? 

- ### The month with the most net generation is Janurary, while April is the month with the least net output
- ### Net energy output has been ever increasing from 1973 to 1999
- ### Net output becomes fixated producing from 700000 to 800000 from 1999 to 2019 (cause?) 

### Intial Predictions: 
- ### Net output has been increasing from 1973 to 1999 due to new power plants being built
- ### It becomes fixated as no new power plants were being built
- ### It becomes fixated as the current running power plants are of the same models or older, only being able to produce a certain amount each month

### As we can see from the data above represented as a pie chart it simply showcases the generation distribution from 2013-2019 nothing regarding which exact power plant produces the most or least, for now the analysis will compare the United States data compared to the other nuclear powered countries

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of countries to compare with the USA
selected_countries = [
    'United States of America', 'China', 'Russia', 'North Korea', 
    'France', 'United Kingdom', 'India', 'Germany', 'Ukraine', 'Canada', 'Mexico', 'Finland', 'Slovakia' , 'Czech Republic' 
]

# Filter data for the selected countries
selected_data = us_plants[us_plants['country'].isin(selected_countries)]

# Extract relevant columns for comparison
comparison_columns = [
    'country', 'total_capacity_gw', 'capacity_gw_fuel_nuclear', 'count', 
    'count_generation_gwh_2013', 'count_generation_gwh_2014', 'count_generation_gwh_2015'
]

# Select only the necessary columns
selected_data = selected_data[comparison_columns]

# Plot comparison between selected countries
plt.figure(figsize=(15, 7))

# Plot the total capacity for the selected countries
plt.subplot(1, 2, 1)
plt.bar(selected_data['country'], selected_data['total_capacity_gw'], color='skyblue')
plt.title('Total Capacity (GW) - Major Countries vs USA')
plt.xlabel('Country')
plt.ylabel('Total Capacity (GW)')
plt.xticks(rotation=30, ha='right')

# Plot the nuclear capacity for the selected countries
plt.subplot(1, 2, 2)
plt.bar(selected_data['country'], selected_data['capacity_gw_fuel_nuclear'], color='lightgreen')
plt.title('Nuclear Capacity (GW) - Major Countries vs USA')
plt.xlabel('Country')
plt.ylabel('Nuclear Capacity (GW)')
plt.xticks(rotation=30, ha='right')

# Display the plots
plt.tight_layout()
plt.show()
```

### Before we evaluate the data lets first understand what total capacity and nuclear capacity is regarding within the comparsion:

- ### Total Capacity: Represents the overall energy capacity a country has regarding generating power from various resources such as oil, gas, solar, wind , geothermal, etc.
   
- ### Nuclear Capacity: Represents specifically the power output from nuclear energy that each nuclear powered country is able to produce. Like total capacity it is measured in GigaWatts or MegaWatts (GW) (MW).
